```python
text_file = open("The hungry mouse.txt")
text = text_file.read()
```

# **Summary of the story**
The narrative centers around a mouse who, after enduring prolonged hunger, discovers a basket of corn. Driven by hunger, she eats excessively, becoming too large to exit the basket through the small entry hole. Her predicament highlights the consequences of overindulgence. A rat suggests that the mouse wait and lose weight to escape, emphasizing the lesson that excess can be detrimental. This story underscores the importance of moderation and self-control.


# Import required libraries :

```python
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
import nltk
nltk.download("popular")
```

```python
sentences = sent_tokenize(text)
print(len(sentences))
sentences
```

```python
words = word_tokenize(text)
print (len(words))
words
```

```python
from nltk.probability import FreqDist
fdist = FreqDist(words)
fdist.most_common(15)
```

```python
#plotting 15 common words
import matplotlib.pyplot as plt
fdist.plot(15)
```

```python
#creating a empty list
no_punctuation= []
for w in words:
    if w.isalpha():
        no_punctuation.append(w.lower())
#printing list
print (len(no_punctuation))
no_punctuation
```

```python
fdist = FreqDist(no_punctuation)
fdist.most_common(15)
```

```python
#Plotting most common words on grpah
fdist.plot(15)
```

```python
from nltk.corpus import stopwords
stopwords = stopwords.words("english")
stopwords
```

```python
#creating a empty list remove_words to store other words excluding the stopwords
remove_words = []
for w in no_punctuation:
    if w not in stopwords:
        remove_words.append(w)
#printing other words from the story
print(len(remove_words))
print(remove_words)
```

```python
#printing common words excluding the stopwords
fdist = FreqDist(remove_words)
fdist.most_common(15)
```

```python
#Plotting the 15 most common words on grpah
fdist.plot(15)
```

```python
from wordcloud import WordCloud
#Generating the wordcloud
wordcloud = WordCloud().generate(text)
#plotting the graph
plt.figure(figsize = (15,15))
plt.imshow(wordcloud)
#axis off doesn't show the value of the axis
plt.axis("off")
plt.show()
```

```python
import numpy as np
from PIL import Image
#using square png
char_mask = np.array(Image.open("square.png"))
wordcloud = WordCloud(background_color="black",mask=char_mask).generate(text)

plt.figure(figsize = (30,30))
plt.imshow(wordcloud)

plt.axis("off")
plt.show()
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()

l1 = ["disappoint","disappointing","disappointed"]
for w in l1:
    print(porter.stem(w))
```

```python
from nltk.stem import PorterStemmer
porter = PorterStemmer()
word_list = ["disappoint","leaves","mouse","diet"]
for w in word_list:
    print(porter.stem(w))
```

```python
from nltk import WordNetLemmatizer
lemma = WordNetLemmatizer()
l2 =  ["disappoint","disappointing","disappointed"]
for w in l2:
    print(lemma.lemmatize(w ,pos="v"))
```

```python
from nltk import WordNetLemmatizer
lemma = WordNetLemmatizer()
auxillary_verb = ["am","is","are","was","were"]
for w in auxillary_verb:
    print(lemma.lemmatize(w ,pos="v"))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize('disappointing', pos="v"))
print(lemmatizer.lemmatize('disappointing', pos="n"))
print(lemmatizer.lemmatize('disappointing', pos="a"))
print(lemmatizer.lemmatize('disappointing', pos="r"))
```
